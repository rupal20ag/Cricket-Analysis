```python
import pandas as pd
import numpy as np
import json
```

```python
with open("C:/Users/Tushar/Documents/Data Analysis Project using python/data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_match_results.json") as f:
    data=json.load(f)
len(data)
df_match_summary=pd.DataFrame(data[0]["matchSummary"])
df_match_summary.rename({"scorecard":"Scorecard_ID"},axis=1,inplace=True)
df_match_summary
```

```python
with open("C:/Users/Tushar/Documents/Data Analysis Project using python/data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_batting_summary.json") as f:
    data=json.load(f)
    all_rec=[]
    for rec in data:
        all_rec.extend(rec["battingSummary"])
all_rec
df_batting_summary=pd.DataFrame(all_rec)
df_batting_summary
```

```python
score_iddict={}
for index,row in df_match_summary.iterrows():
    key1=row["team1"]+" Vs "+row["team2"]
    key2=row["team2"]+" Vs "+row["team1"]
    score_iddict[key1]=row["Scorecard_ID"]
    score_iddict[key2]=row["Scorecard_ID"]
score_iddict
```

```python
with open("C:/Users/Tushar/Documents/Data Analysis Project using python/data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_bowling_summary.json") as f:
    data=json.load(f)
    all_rec=[]
    for rec in data:
        all_rec.extend(rec["bowlingSummary"])
all_rec

df_bowling_summary=pd.DataFrame(all_rec)

df_bowling_summary["Score_id"]=df_bowling_summary["match"].map(score_iddict)

sp_char=df_bowling_summary["bowlerName"].str.contains(r"[^a-zA-Z\s]",regex=True)
df_bowling_summary.loc[sp_char,"bowlerName"]=df_bowling_summary.loc[sp_char,"bowlerName"].str.replace(r"[^a-zA-Z\s]","",regex=True)
df_bowling_summary
```

```python
with open("C:/Users/Tushar/Documents/Data Analysis Project using python/data-analytics-project-for-beginners/t20_json_files/t20_json_files/t20_wc_player_info.json") as f:
    data=json.load(f)
data
df_player_info=pd.DataFrame(data)

sp_char=df_player_info["name"].str.contains(r"[^a-zA-Z\s]",regex=True)
df_player_info.loc[sp_char,"name"]=df_player_info.loc[sp_char,"name"].str.replace(r"[^a-zA-Z\s]","",regex=True)
df_player_info
```

```python
df_batting_summary["Out/Not Out"]=np.where(df_batting_summary["dismissal"] == "","Not Out","Out") 
df_batting_summary.head(10)
```

```python
df_batting_summary.drop(["dismissal"],axis=1,inplace=True)
df_batting_summary
```

```python
special_char=df_batting_summary["batsmanName"].str.contains(r"[^a-zA-Z\s]", regex=True) #this shows whether in the batsmanName col special char exist or not
df_batting_summary.loc[special_char,"batsmanName"]=df_batting_summary.loc[special_char,"batsmanName"].str.replace(r"[^a-zA-Z\s]","",regex=True)
df_batting_summary.head(20)
```

```python
df_batting_summary["Score_id"]=df_batting_summary["match"].map(score_iddict)
df_batting_summary
```
